```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Let's take a look at life expectancy vs. GDP in the US over the years

```python
us_lifeexp_gdp_df = lifeexp_gdp_df[lifeexp_gdp_df.Country == 'United States of America'].reset_index(drop=True)
```

```python
print(us_lifeexp_gdp_df.head())
```

```python
print(us_lifeexp_gdp_df.info())
```

```python
fig, ax1 = plt.subplots()
plt.title("United States")
color = 'tab:red'
ax1.set_xlabel('Years')
ax1.set_ylabel('GDP')
ax1.plot(us_lifeexp_gdp_df.Year, us_lifeexp_gdp_df.GDP, label='GDP', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend()
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Life Expectancy', color=color)
ax2.plot(us_lifeexp_gdp_df.Year, us_lifeexp_gdp_df['Life expectancy at birth (years)'], label='Life Expectancy at Birth (Years)', color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.legend(loc=1)
fig.tight_layout()
plt.show()
```

From the graph, we can conclude that, for the United States at least, GPD and Life Expectancy at Birth is tightly and positively correlated

Now let's see if gains in life expectancy and GDP are comparable across all countries

```python
print(lifeexp_gdp_df.head())
```

```python
GDP_by_year = lifeexp_gdp_df.groupby('Year').GDP.sum().reset_index()
```

```python
print(GDP_by_year)
```

```python
lifeexp_by_year = lifeexp_gdp_df.groupby('Year')['Life expectancy at birth (years)'].mean().reset_index()
```

```python
print(lifeexp_by_year)
```

```python
lifeexp_vs_GDP = GDP_by_year.merge(lifeexp_by_year)
print(lifeexp_vs_GDP)
```

```python
fig, ax1 = plt.subplots()
plt.title("Global GDP vs. Mean Life Expectancy at Birth")
color = 'tab:red'
ax1.set_xlabel('Years')
ax1.set_ylabel('GDP')
ax1.plot(lifeexp_vs_GDP.Year, lifeexp_vs_GDP.GDP, label='GDP', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend()
ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('Life Expectancy', color=color)
ax2.plot(lifeexp_vs_GDP.Year, lifeexp_vs_GDP['Life expectancy at birth (years)'], label='Life Expectancy at Birth (Years)', color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.legend(loc=1)
fig.tight_layout()
plt.show()
```

Globally, GDP growth outpaced Life-expectancy gains, the two trends only coming close after the GFC of 2008

Let's see if rate of change of GDP over 2002 to 2014 and rate of change of life expectancy is related.

```python
lifeexp_vs_GDP['GDP ROC'] = lifeexp_vs_GDP.GDP.diff()
```

```python
lifeexp_vs_GDP['Life Expectancy at Birth ROC'] = lifeexp_vs_GDP['Life expectancy at birth (years)'].diff()
```

```python
print(lifeexp_vs_GDP.dropna())
```

```python
fig, ax1 = plt.subplots()
plt.title("Global GDP Rate of Chanage vs. Mean Life Expectancy at Birth Rate of Change")
color = 'tab:red'
ax1.set_xlabel('Mean Years Change in Years')
ax1.set_ylabel('GDP Change in Trillions')
ax1.set_xticks([2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015])
ax1.set_xticklabels(["From {} to {}".format(year-1, year) for year in range(2001, 2016, 1)], rotation = 90)
ax1.plot(lifeexp_vs_GDP.Year, lifeexp_vs_GDP['GDP ROC'], label='GDP', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend()
ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('Life Expectancy', color=color)
ax2.plot(lifeexp_vs_GDP.Year, lifeexp_vs_GDP['Life Expectancy at Birth ROC'], label='Life Expectancy at Birth ROC', color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.legend(loc=1)
fig.tight_layout()
plt.show()
```

```python
print(lifeexp_vs_GDP.head())
```

```python
lifeexp_vs_GDP['GDP pct change'] = lifeexp_vs_GDP.GDP.pct_change()*100
```

```python
lifeexp_vs_GDP['Mean Life Expectancy at Birth pct Change'] = lifeexp_vs_GDP['Life expectancy at birth (years)'].pct_change()*100
```

```python
print(lifeexp_vs_GDP)
```

```python
fig, ax1 = plt.subplots()
plt.title("Global GDP vs. Mean Life Expectancy at Birth % Change")
color = 'tab:red'
ax1.set_xlabel('Mean Years % Change')
ax1.set_ylabel('GDP % Change')
ax1.set_xticks([2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015])
ax1.set_xticklabels(["From {} to {}".format(year-1, year) for year in range(2001, 2016, 1)], rotation = 90)
ax1.set_yticks([i for i in range(-3, 11, 1)])
ax1.set_yticklabels([f'{i}%' for i in range(-3, 11, 1)])
ax1.plot(lifeexp_vs_GDP.Year, lifeexp_vs_GDP['GDP pct change'], label='GDP % Change', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend()



ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('Mean Life Expectancy % Change', color=color)
ax2.plot(lifeexp_vs_GDP.Year, lifeexp_vs_GDP['Mean Life Expectancy at Birth pct Change'], label='Life Expectancy at Birth % Change', color=color)
ax2.set_yticks([i/10 for i in range(1, 11, 1)])
ax2.set_yticklabels([f'0.{i}%' for i in range(1, 11, 1)])
ax2.tick_params(axis='y', labelcolor=color)
plt.legend(loc=1)
fig.tight_layout()
plt.show()
```

Based on the pct change in GDP and Life Expectancy, throughout the period from 2000 to 2015, the change in mean life expectancy across the globe accelerated when GDP accelerated with more rapid acceleration occuring when GDP's rate of increase accelerated, then stagnated but didn't fall when GDP contracted from 2006 to 2009, recovered its rise from 2010 with one year lag behind GDP, and continue its rapid rise until 2011, when the GDP contracted once again and the rate of incrase slowed down.
